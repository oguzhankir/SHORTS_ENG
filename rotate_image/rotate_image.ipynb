```python
from PIL import Image
```

```python
image_path = r'C:\Users\maest\Desktop\image.png'
Image.open(image_path)
```

```python
def rotate(image_path, degree, output_path):
    image_obj = Image.open(image_path)
    rotated_image = image_obj.rotate(degree)
    rotated_image.save(output_path)

output_path = r'C:\Users\maest\Desktop\rotated_image.png'

rotate(image_path, 180, output_path)
```

```python
Image.open(output_path)
```
