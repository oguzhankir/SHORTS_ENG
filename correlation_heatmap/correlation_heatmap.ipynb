```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Users\maest\Downloads\concrete.csv")
```

```python
def plot_heatmap(df):
    cmap = sns.diverging_palette(250, 15, s=75, 
                                 l=40, n=9, 
                                 center="light", 
                                 as_cmap=True)


    matrix = df[df.columns].corr(method="pearson")
    
    mask = np.triu(np.ones_like(matrix, dtype=bool))
    fig, ax = plt.subplots(figsize=(7, 10))
    
    sns.heatmap(matrix, 
                mask=mask, 
                cmap=cmap, 
                square=True, 
                annot=True, 
                fmt=".2f", 
                ax=ax)
    
    return 


plot_heatmap(df)
```
